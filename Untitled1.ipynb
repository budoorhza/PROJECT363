```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
```

```python
data_path = "/content/data"  # المسار الذي يحتوي على بيانات PlantVillage
```

```python
# إعداد مولد بيانات مع التطبيع وتقسيم البيانات
datagen = ImageDataGenerator(
    rescale=1.0/255,  # تطبيع القيم إلى [0, 1]
    validation_split=0.2  # تقسيم 20% من البيانات للتحقق
)

# تحميل بيانات التدريب
train_generator = datagen.flow_from_directory(
    data_path,
    target_size=(128, 128),  # حجم الصور بعد التغيير
    batch_size=32,          # حجم الدفعة
    class_mode='categorical',  # لأن المهمة تصنيف
    subset='training'        # جزء التدريب
)

# تحميل بيانات التحقق
val_generator = datagen.flow_from_directory(
    data_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    subset='validation'      # جزء التحقق
)
```

```python
# بناء نموذج بسيط باستخدام الطبقات التلافيفية
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(train_generator.class_indices), activation='softmax')  # عدد الفئات
])

# تجميع النموذج
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()  # عرض ملخص النموذج
```

```python
# تدريب النموذج
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=10,  # عدد الدورات
    steps_per_epoch=train_generator.samples // 32,
    validation_steps=val_generator.samples // 32
)
```

```python
# رسم دقة النموذج
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# رسم الخسارة
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# الحصول على صورة من بيانات التحقق
sample_image, sample_label = next(val_generator)

# توقع الفئة
predictions = model.predict(sample_image)
predicted_labels = np.argmax(predictions, axis=1)

# عرض الصورة مع التسمية الحقيقية والتنبؤ
plt.figure(figsize=(12, 8))
for i in range(6):  # عرض أول 6 صور
    plt.subplot(2, 3, i+1)
    plt.imshow(sample_image[i])
    true_label = list(val_generator.class_indices.keys())[np.argmax(sample_label[i])]
    predicted_label = list(val_generator.class_indices.keys())[predicted_labels[i]]
    plt.title(f"True: {true_label}\nPred: {predicted_label}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```
